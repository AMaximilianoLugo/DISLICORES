```python

# Limpieza de datos (siguiendo los pasos originales)

# Rellenar valores nulos en 'Price' con la media
purchase_prices['Price'].fillna(purchase_prices['Price'].mean(), inplace=True)

# Rellenar valores nulos en 'PurchasePrice' con la media
purchase_prices['PurchasePrice'].fillna(purchase_prices['PurchasePrice'].mean(), inplace=True)

# Eliminar duplicados
purchase_prices.drop_duplicates(inplace=True)

# Guardar el DataFrame filtrado en un archivo CSV
purchase_prices.to_csv('PreciosCompra2017.csv', index=False)

print("\nLimpieza de datos completada y guardada.")

```

```python
# Paso 1: Extraer los valores numéricos de la columna 'Size' (eliminar las unidades 'mL')
# Usamos una raw string para la expresión regular
purchase_prices['Size_numeric'] = purchase_prices['Size'].str.extract(r'(\d+)').astype(float)

# Asegurarnos de que la columna 'Volume' sea numérica
purchase_prices['Volume'] = pd.to_numeric(purchase_prices['Volume'], errors='coerce')

# Paso 2: Verificar si los valores de 'Size_numeric' coinciden con 'Volume'
coinciden = (purchase_prices['Size_numeric'] == purchase_prices['Volume']).all()

if coinciden:
    # Si los valores coinciden, eliminar la columna 'Size'
    purchase_prices.drop(columns=['Size', 'Size_numeric'], inplace=True)
    print("\nTodos los valores de 'Size_numeric' coinciden con 'Volume'. La columna 'Size' ha sido eliminada.")
else:
    print("\nLos valores de 'Size_numeric' no coinciden completamente con 'Volume'. No se eliminará ninguna columna.")

# Guardar el DataFrame actualizado
purchase_prices.to_csv('PreciosCompra2017_actualizado.csv', index=False)

print("\nProceso completado. El archivo actualizado ha sido guardado.")
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset limpio
df = pd.read_csv('PreciosCompra2017.csv')
```

```python

# Configurar el estilo de las visualizaciones
sns.set(style="whitegrid")

# Visualización 1: Distribución de los Precios de Compra
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True, color='blue')
plt.title('Distribución de los Precios de Compra', fontsize=16)
plt.xlabel('Precio', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.xlim(0, 500) 
plt.show()


# Conclusión 1: Observar la distribución de los precios. 
# Podemos ver los precios se concentra entre los 0-200 dolares 
```

```python
# Visualización 2: Relación entre el Tamaño y el Precio
plt.figure(figsize=(12, 6)) 
sns.scatterplot(x='Volume', y='Price', data=df, hue='BrandID', palette='viridis')
plt.xticks(rotation=45, ha='right') 
plt.gcf().subplots_adjust(bottom=0.25) 
plt.title('Relación entre Tamaño y Precio', fontsize=16)
plt.xlabel('Tamaño', fontsize=12)
plt.ylabel('Precio', fontsize=12)
plt.show()


# Conclusión 2: no hay relacion entre precio y Tamaño
```

```python
# Visualización 3: Clasificación de Productos y sus Precios
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='Price', data=df)
plt.title('Precios por Clasificación de Producto', fontsize=16)
plt.xlabel('Clasificación', fontsize=12)
plt.ylabel('Precio', fontsize=12)
plt.show()

# Conclusión 3: la clasificacion 1 tiene mayor y dispersion rango de precios 
```

```python

# Filtrar solo las columnas numéricas para la correlación
df_numeric = df.select_dtypes(include='number')

# Visualización: Matriz de Correlación entre Variables Numéricas
plt.figure(figsize=(8, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables', fontsize=16)
plt.show()

# nada tiene correlaciones fuertes. 
```
